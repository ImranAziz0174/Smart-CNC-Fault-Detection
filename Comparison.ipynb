```python
import matplotlib.pyplot as plt
import numpy as np

# ======================== DATA FOR COMPARISON ========================

# Tool Wear Prediction Metrics
logreg_tool = {'accuracy': 0.33, 'precision': 0.20, 'recall': 0.33, 'f1-score': 0.25}
xgb_tool = {'accuracy': 0.60, 'precision': 0.36, 'recall': 0.60, 'f1-score': 0.45}

# Clamping Fault Prediction Metrics
logreg_clamp = {'accuracy': 0.80, 'precision': 0.90, 'recall': 0.80, 'f1-score': 0.82}
xgb_clamp = {'accuracy': 0.80, 'precision': 0.64, 'recall': 0.80, 'f1-score': 0.71}

# Metrics to plot
metrics = ['accuracy', 'precision', 'recall', 'f1-score']

# ======================== PLOT TOOL WEAR ========================

fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(metrics))
width = 0.35

logreg_scores = [logreg_tool[m] for m in metrics]
xgb_scores = [xgb_tool[m] for m in metrics]

rects1 = ax.bar(x - width/2, logreg_scores, width, label='Logistic Regression', color='skyblue')
rects2 = ax.bar(x + width/2, xgb_scores, width, label='XGBoost', color='lightcoral')

ax.set_ylabel('Scores')
ax.set_title('Tool Wear Prediction: Logistic Regression vs XGBoost')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)

for rects in [rects1, rects2]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.show()

# ======================== PLOT CLAMPING FAULT ========================

fig, ax = plt.subplots(figsize=(10, 6))
logreg_scores = [logreg_clamp[m] for m in metrics]
xgb_scores = [xgb_clamp[m] for m in metrics]

rects1 = ax.bar(x - width/2, logreg_scores, width, label='Logistic Regression', color='lightgreen')
rects2 = ax.bar(x + width/2, xgb_scores, width, label='XGBoost', color='orange')

ax.set_ylabel('Scores')
ax.set_title('Clamping Fault Prediction: Logistic Regression vs XGBoost')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)

for rects in [rects1, rects2]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.show()
```
